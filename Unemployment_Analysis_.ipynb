```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
from google.colab import files
import io
```

```python
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['Unemployment in India.csv']))
```

```python
df.head()
```

```python
df.columns = [col.strip().replace(" ", "_").lower() for col in df.columns]
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.rename(columns={
    'estimated_unemployment_rate_(%)': 'unemployment_rate',
    'estimated_employed': 'employed',
    'estimated_labour_participation_rate_(%)': 'labour_participation_rate'}, inplace=True)
numeric_cols = ['unemployment_rate', 'employed', 'labour_participation_rate']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=numeric_cols + ['date'], inplace=True)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['month_num'] = df['date'].dt.month
df['quarter'] = df['date'].dt.to_period('Q')
```

```python
national = df.groupby(['year', 'month_num', 'month'])[numeric_cols].mean().reset_index().sort_values(['year', 'month_num'])
plt.figure(figsize=(14, 6))
sns.lineplot(data=national, x='month', y='unemployment_rate', hue='year', marker='o')
plt.title('National Monthly Unemployment Rate (by Year)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
covid_df = df[df['year'] >= 2020].groupby('year')[numeric_cols].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='year', y='unemployment_rate', data=covid_df)
plt.title("Average Unemployment Rate During Covid Era (2020–onward)")
plt.ylabel("Unemployment Rate (%)")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='region', y='unemployment_rate', data=df)
plt.xticks(rotation=90)
plt.title("Unemployment Rate Distribution by Region")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
top5 = df.groupby('region')['unemployment_rate'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(6,4))
top5.plot(kind='barh', color='tomato')
plt.title("Top 5 States with Highest Avg Unemployment")
plt.xlabel("Unemployment Rate (%)")
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
heatmap_data = df.pivot_table(values='unemployment_rate', index='region', columns='month_num', aggfunc='mean')

plt.figure(figsize=(16,8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", linewidths=0.5)
plt.title("Heatmap of Average Monthly Unemployment Rate by Region")
plt.xlabel("Month Number")
plt.ylabel("Region")
plt.show()
```

```python
seasonal = df.groupby(['month_num', 'month'])[numeric_cols].mean().reset_index().sort_values('month_num')

plt.figure(figsize=(10, 5))
sns.lineplot(data=seasonal, x='month', y='unemployment_rate', marker='o', sort=False)
plt.title('Seasonal Unemployment Pattern in India')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=seasonal, x='month', y='employed', palette='coolwarm')
plt.title("Avg Employment per Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
